```python
ngc_2682 = pd.read_csv('stab213_supplemental_files/Gaia_cat.txt')
```

```python
ngc_2682 = ngc_2682[ngc_2682['cluster']=="NGC_2682"].reset_index()
```

```python
ngc_2682
```

```python
cat = ngc_2682
```

```python
co = 0.7
```

```python
qf_0 = cat[cat['qf']==0]
cand_0 = qf_0[(qf_0['P_F6']>=co) & (qf_0['P_F10']<co)]
field_0 = qf_0[(qf_0['P_F6']<co) | (qf_0['P_F10']>=co)]
qf_1 = cat[cat['qf']==1]
cand_1 = qf_1[(qf_1['P_F6']>=co) & (qf_1['P_F10']<co)]
field_1 = qf_1[(qf_1['P_F6']<co) | (qf_1['P_F10']>=co)]
cand = pd.concat([cand_0, cand_1], axis=0)

cat_M = cat[cat['class']=='M']
cat_C = cat[cat['class']=='C']

source_ids_1 = cand['source_id'].astype(str)
g_mag_1 = cand['g_mag'].round(4).astype(str)
g_rp_1 = cand['g_rp'].round(4).astype(str)
source_ids_2 = cat_C['source_id'].astype(str)
g_mag_2 = cat_C['g_mag'].round(4).astype(str)
g_rp_2 = cat_C['g_rp'].round(4).astype(str)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2,
                    horizontal_spacing=0.12,subplot_titles=("Comparision of P<sub>F6</sub> and P<sub>F10</sub>", "CMD of NGC 2682 (based on membership criteria)"))

fig.add_trace(go.Scatter(x = cand['P_F6'], y = cand['P_F10'], mode="markers", marker=dict(color='green',
                                                                                            symbol='circle-open',
                                                                            size=7), name='remove',
                         hoverinfo='text',hovertext = '<b>'+"Source ID: " + 
                         source_ids_1 +'</b>'+"<br>" + 'G[mag]: '+g_mag_1 + "<br>" + 'G - '+ 'G'+'<sub>'+'RP'+'</sub>'+ ': ' + g_rp_1, 
                         showlegend=False,
                        hoverlabel=dict(font=dict(family="Times New Roman"))), row=1, col=1)

fig.add_trace(go.Scatter(x = field_0['P_F6'], y = field_0['P_F10'], mode="markers", marker=dict(color='red',
                                                                                            symbol='circle-open',
                                                                            size=7), name='qf = 0',hoverinfo='skip'), row=1, col=1)


fig.add_trace(go.Scatter(x = field_1['P_F6'], y = field_1['P_F10'], mode="markers", marker=dict(color='black',
                                                                                            symbol='circle',
                                                                            size=4), name='qf = 1',hoverinfo='skip'), row=1, col=1)

fig.add_trace(go.Scatter(x=cat_M['g_rp'], y = cat_M['g_mag'], mode="markers", marker=dict(color='grey',size=2), name='Members', hoverinfo='skip'),row=1, col=2)
fig.add_trace(go.Scatter(x=cat_C['g_rp'], y = cat_C['g_mag'], mode="markers", marker=dict(color='green',
                                                                                          symbol="circle-open",size=7), name='Candidates',
                        hoverinfo='text',hovertext = '<b>'+"Source ID: " + 
                         source_ids_2 +'</b>'+"<br>" + 'G[mag]: ' + g_mag_2 + "<br>" + 'G - '+'G'+'<sub>'+'RP'+'</sub>'+ ': ' + g_rp_2, 
                       hoverlabel=dict(font=dict(family="Times New Roman")) ),row=1, col=2)

fig.add_hline(y=co, row=1, col=1)
fig.add_vline(x=co, row=1, col=1)

fig.update_xaxes(title_text=r"$\rm P_{F6}$", row=1, col=1)
fig.update_yaxes(title_text=r"$\rm P_{F10}$", row=1, col=1)
fig.update_xaxes(title_text=r"$\rm G-G_{RP}$", row=1, col=2)
fig.update_yaxes(title_text=r'$\rm G[mag]$', row=1, col=2)

fig.update_yaxes(autorange="reversed",row=1, col=2)

fig.update_layout(height=800, width=1400)
fig.update_traces(showlegend=False, selector=dict(type='scatter'),row=1, col=2)

fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.01,font = dict(family="Times New Roman", size=10)
))

fig.update_annotations(font_family="Times New Roman", font_size=20)

plotly.offline.plot(fig, filename='a_b.html',auto_open=False,include_mathjax='cdn')
fig.write_image('a_b_gaia.png')

# fig.update_layout(title=dict(text="Comparision of P_F6 and P_F10 and CMD of NGC 2682 (based on membership criteria)",xanchor= 'center',
#         yanchor= 'top',y=0.92, x=0.5),font=dict(
#         family="Times New Roman",
#         size=15)#,row=1,col=1)
fig.show()
```

```python
f_stars = cat[(cat['P_F6'] < co) & (cat['P_F8'] < co)]
m_c_stars = cat[(cat['P_F6'] >= co) | (cat['P_F8'] >= co)]

cat_wo_F = cat[cat['class']!='F']
cat_F = cat[cat['class']=='F']

peculiarity_1 = (m_c_stars['P_F6']-m_c_stars['P_F8']).round(3).astype(str)
gmag_1 = m_c_stars['g_mag'].round(4).astype(str)
grp_1 = m_c_stars['g_rp'].round(4).astype(str)
source_1 = m_c_stars['source_id'].astype(str)

peculiarity_2 = (cat_wo_F['P_F6']-cat_wo_F['P_F8']).round(3).astype(str)
gmag_2 = cat_wo_F['g_mag'].round(4).astype(str)
grp_2 = cat_wo_F['g_rp'].round(4).astype(str)
source_2 = cat_wo_F['source_id'].astype(str)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# fig = go.Figure()
fig = make_subplots(rows=1, cols=2,
                    horizontal_spacing=0.12,subplot_titles=("Comparision of P<sub>F6</sub> and P<sub>F8</sub>", "CMD of stars according to Peculiarity (P<sub>F6</sub> - P<sub>F8</sub>)")
                   )

fig.add_trace(go.Scatter(x=f_stars['P_F6'], y=f_stars['P_F8'], mode="markers", marker=dict(color='grey', size=2), 
#                          name=r"$\rm P_{F8} \; AND \; P_{F6} < co$",
                         name="P_F8 AND P_F6 < co",
                        hoverinfo='skip'), row=1, col=1)
fig.add_trace(go.Scatter(x=m_c_stars['P_F6'], y=m_c_stars['P_F8'], mode="markers", 
                         marker=dict(color=m_c_stars['P_F6']-m_c_stars['P_F8'],cmax=0.4, cmin=-0.4, 
                         colorbar=dict(
                             title=dict(
#                                  text = r"$\rm P_{F6} \; - \; P_{F8}$", 
#                                      text = "P_F6 - P_F8",
                                        text = "Peculiarity",
                                        side = "right", 
                                        font=dict(family = "Times New Roman",size=20)),
                             xpad=0, 
                             tickfont=dict(family="Times New Roman")), colorscale="Portland"), 
#                          name=r"$\rm P_{F8} \; OR \; P_{F6} > co$",
                         name="P_F8 AND P_F6 < co",
                         hoverinfo='text',hovertext = '<b>'+"Source ID: " + 
                         source_1 +'</b>'+"<br>" + 'G[mag]: ' + gmag_1 + "<br>" + 'G - '+'G'+'<sub>'+'RP'+'</sub>'+ ': ' + grp_1
                         + '<br>'+ 'Peculiarity: '+ peculiarity_1, 
                       hoverlabel=dict(font=dict(family="Times New Roman"))
                        ), row=1, col=1)

fig.add_trace(go.Scatter(x=cat_wo_F['g_rp'], y = cat_wo_F['g_mag'], mode="markers", 
                         marker=dict(color=cat_wo_F['P_F6']-cat_wo_F['P_F8'],cmax=0.4, cmin=-0.4, 
#                                      colorbar=dict(title=dict(text="P_F6 - P_F8",
#                                                              side="right", font=dict(family = "Times New Roman",size=17)), 
#                                                    tickfont=dict(family="Times New Roman"),
#                                                   len=0.55,y=0.25), 
                                     colorscale="Portland"
                                    ),
                                              hoverinfo='text',hovertext = '<b>'+"Source ID: " + 
                         source_2 +'</b>'+"<br>" + 'G[mag]: ' + gmag_2 + "<br>" + 'G - '+'G'+'<sub>'+'RP'+'</sub>'+ ': ' + grp_2
                         + '<br>'+ 'Peculiarity: '+ peculiarity_2, 
                       hoverlabel=dict(font=dict(family="Times New Roman"))
                        ),row=1, col=2)

fig.add_trace(go.Scatter(x=cat_F['g_rp'], y = cat_F['g_mag'], mode="markers", hoverinfo='skip',marker=dict(color='grey', size=2)),row=1, col=2)

fig.update_yaxes(autorange="reversed", row=1, col=2)

fig.add_hline(y=co, row=1, col=1)
fig.add_vline(x=co, row=1, col=1)

fig.update_xaxes(title_text=r"$\rm P_{F6}$", row=1, col=1)
fig.update_yaxes(title_text=r"$\rm P_{F8}$", row=1, col=1)
fig.update_xaxes(title_text=r"$\rm G-G_{RP}$", row=1, col=2)
fig.update_yaxes(title_text=r'$\rm G[mag]$', row=1, col=2)

fig.update_layout(height=800, width=1400)
fig.update_traces(showlegend=False, selector=dict(type='scatter'), row=1, col=2)
fig.update_traces(showlegend=False, selector=dict(type='scatter'), row=1, col=1)

fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.01, font = dict(family="Times New Roman", size=12)
))

fig.update_annotations(font_family="Times New Roman", font_size=20)


plotly.offline.plot(fig, filename='c_d.html',auto_open=False,include_mathjax='cdn')
fig.write_image('c_d_gaia.png')

fig.show()
```
